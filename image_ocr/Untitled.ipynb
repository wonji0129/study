# image -> pdf (PIL)

```python
from PIL import Image

img = Image.open("./data/vincent_m65_fishtail_parka_002_size.jpg")
im = img.convert('RGB')

im.save('size_info.pdf')
```

# PyPDF2

```python
import PyPDF2

file = open("size_info.pdf", 'rb')
fileReader = PyPDF2.PdfFileReader(file)
 
```

```python
# 첫 번째 페이지 정보를 가져온다
pageObj = fileReader.getPage(0)
pageObj
```

```python
pageObj.extractText()
```
